```python
import pandas as pd
import os
```

```python
dataset = 'nicla_digits'
```

```python
folder = './source'
target = 'digit'
```

```python
result = pd.DataFrame()
for file in sorted(os.listdir(folder)):
  class_ = int((file.split(sep='.')[0]))
  df = pd.read_csv(os.path.join(folder, file), skip_blank_lines=True)
  df[target] = class_
  result = pd.concat((result, df), axis=0, ignore_index=True)
```

```python
result.drop(columns=['gyro_X', 'gyro_Y', 'gyro_Z'], inplace=True)
```

```python
result.to_csv(f'{dataset}.csv', index=False)
```
